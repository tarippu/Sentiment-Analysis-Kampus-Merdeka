# Lowercase

```python
# Making Everything in LowerCase No Repeatations
df['Tweets'] = df['Tweets'].apply (lambda x: " ".join (x.lower () for x in x.split ()))
df['Tweets']
```

```python
df['Tweets'] = df['Tweets'].str.replace('[^\w\s]','',regex=True)
df['Tweets'].head()
#REMOVING THE PUNCTUCATION
```

```python
#Cleaning
def remove_punct(Tweets):
    Tweets = re.sub(r'[^a-zA-z0-9]', ' ', str(Tweets))
    Tweets = re.sub(r'\b\w{1,2}\b', '', Tweets)
    Tweets = re.sub(r'\s\s+', ' ', Tweets)
    return Tweets

df['Tweets'] = df['Tweets'].apply(lambda x: remove_punct(x))
df.head(10)
```

```python
#Tokenization
def tokenization_text(text):
    text = re.split('\W+',text)
    return text

df['TOKENIZATION'] = df['Tweets'].apply(lambda x: tokenization_text(x.lower()))
df.head(10)
```

```python
#Stop Removal

stopword = nltk.corpus.stopwords.words('indonesian')
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text

df['STOP_REMOVAL'] = df['TOKENIZATION'].apply(lambda x: remove_stopwords(x))
df.head(5)
```

# Labeling

```python
df['Label'] = ''
for i,x in df.STOP_REMOVAL.iteritems():
    label = TextBlob(x)
    df['Label'][i] = label.sentiment.polarity
    print("Index: ", i ,  "Label" , label.sentiment.polarity)
```

```python
def polarity_to_label(x):
    if(x >= -1 and x < 0):
        return 'negative'
    if(x == 0):
        return 'neutral'
    if(x > 0 and x <= 1):
        return 'positif'
df.Label = df.Label.apply(polarity_to_label)
```
